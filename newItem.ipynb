## 欧式距离
对应维度的平方和，开平方根

## 曼哈顿距离，城市街区距离
对应维度的绝对值和，求和

## 切比雪夫距离
在曼哈顿距离的基础上，加上可以斜着走
对应维度的绝对值，求最大值

## 闵可夫斯基距离
不是一种新的距离的度量方式
是对多个距离度量公式的概括性的表达


## 特征预处理

    归一化：通过对原始数据进行变换，把数据的值映射到【min, max】（默认[0, 1]）之间
        可能受到特殊值的影响
        通过公式把 各列的值映射到 【0, 1】之间
        x' = (当前值 - 该列最小值) / (该列最大值 - 该列最小值)
        x'' = x' * (max - min) + min
    弊端：易受到最大值和最小值的影响，一般用于处理 小数据集

    标准化：通过对原始数据进行变换，为均值为0 标准差为1 的标志正态分布的数据
        通过公式把 各列的值映射到 【-1, 1】之间
        x' = (当前值 - 该列均值) / 该列标准差
        x'' = x' * 2 + 1

    无论是归一化还是标准化，目的都是为了解决 模型量纲（单位）问题，导致模型评估低等问题

```python
## 鸢尾花数据集
from sklearn.datasets import load_iris
import pandas as pd
##根据 Seaborn 绘图
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split    # 数据集划分
from sklearn.neighbors import KNeighborsClassifier      # KNN分类器
from sklearn.preprocessing import StandardScaler        # 数据标准化
from sklearn.metrics import accuracy_score              # 模型评估，计算模型预测的准确性

# 数据加载
def dm01_load_iris():
    iris_data = load_iris()

    # print(f'数据集：{iris_data}')
    print(f'数据集：{iris_data.data[:5]}')
    print(f'标签集：{iris_data.target[:5]}')
    print(f'名称集：{iris_data.target_names}')
    print(f'特征集：{iris_data.feature_names}')
    # print(f'数据集描述：{iris_data.DESCR}')


# 数据可视化
def dm02_iris_data_visualization():
    iris_data = load_iris()
    ##  数据可视化
    iris_df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
    iris_df['label'] = iris_data.target
    # iris_data.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'label']
    # print(iris_df)
    
    sns.lmplot(x='sepal length (cm)', y='sepal width (cm)', hue='label', data=iris_df, fit_reg=False)
    plt.title('iris data')
    # plt.tight_layout()
    plt.show()


# 数据集划分
def dm03_split_train_test():
    iris_data = load_iris()
    # 从150个数据中150个特征中随机选择120个数据作为训练集，30个数据作为测试集
    x_train, x_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, test_size=0.2, random_state=13)
    print(f'训练集特征：{x_train},  训练集：{x_train.shape}， 个数：{len(x_train)}')
    print(f'测试集特征：{x_test}, 测试集{x_test.shape}， 个数：{len(x_test)}')
    
def dm04_iris_evaluate_test():
    # 加载数据
    iris_data = load_iris()
    # 数据预处理
    x_train, x_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, test_size=0.2, random_state=13)
    # 特征工程
    # 创建标准化对象
    transfer = StandardScaler()

    # fit_transform 兼具训练和转换的功能
    x_train= transfer.fit_transform(x_train)
    # transform 纯转换
    x_test = transfer.transform(x_test)


    # 模型训练
    estimator = KNeighborsClassifier(n_neighbors=3)
    estimator.fit(x_train, y_train)
    # 模型预测
    y_predict = estimator.predict(x_test)
    print(f'预测结果：{y_predict}')


    # 模型评估
    # 直接评分  基于：训练集的特征 和 训练集的标签
    print(f'真实结果：{estimator.score(x_test, y_test)}')
    # 基于：测试集的标签 和 预测的标签
    print(f'准确率：{accuracy_score(y_test, y_predict)}')





# dm01_load_iris()
# dm02_iris_data_visualization()
# dm03_split_train_test()

dm04_iris_evaluate_test()
```

**线性回归**

```python
# 机器学习基础模型
# 基础线性回归算法优化

import random
import torch
import torch.nn as nn
import numpy as np
# from d2l import torch as d2l
import d2l

# 根据带有噪声的线性模型构造数据集。设数据集大小为n，批量大小为batch_size。
# 生成批量数据
def synthetic_data(w, b, num_examples):
    """生成 y = Xw + b + 噪声"""
    # 生成一个形状为(num_examples, len(w))的随机正态分布张量，均值为0，标准差为1
    # 这代表特征矩阵X，每一行是一个样本，每一列是一个特征
    X = torch.normal(0, 1, (num_examples, len(w)))
    # 计算线性模型的基本预测值：X与权重w相乘再加上偏置b
    y = torch.matmul(X, w) + b
    # 添加高斯噪声，模拟真实数据中的不确定性
    # 噪声的标准差为0.01，使数据更接近实际情况
    y += torch.normal(0, 0.01, y.shape)
    # 将y重新塑形为(-1, 1)，即保持行数不变，列数变为1
    # 这样确保标签y是一个列向量
    return X, y.reshape((-1, 1))

# 定义真实的权重参数，这是我们要学习的目标参数
# w = [2, -3.4] 表示第一个特征的权重为2，第二个特征的权重为-3.4
true_w = torch.tensor([2, -3.4])
# 定义真实的偏置参数，这是线性模型中的截距项
true_b = 4.2
# 使用定义的参数生成包含1000个样本的人工数据集
# features是特征矩阵，labels是对应的标签
features, labels = synthetic_data(true_w, true_b, 1000)

print('features:', features[0], '\nlabel:', labels[0])

d2l.set_figsize()
d2l.plt.scatter(features[:, 1].detach().numpy(),
                labels.detach().numpy(), 1)


# 定义一个data_iter函数，接受批量大小、特征矩阵和标签向量作为输入
def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))
    # 这些样本是随机读取的，没有特定的顺序
    random.shuffle(indices)
    for i in range(0, num_examples, batch_size):
        batch_indices = torch.tensor(
            indices[i: min(i + batch_size, num_examples)])
        yield features[batch_indices], labels[batch_indices]

bathch_size = 10

for X, y in data_iter(bathch_size, features, labels):
    print(X, '\n', y)
    break

# 定义一个线性模型
def linreg(X, w, b):
    """线性回归模型"""
    return torch.matmul(X, w) + b

# 定义初始化模型参数
w = torch.normal(0, 0.01, size=(2, 1), requires_grad=True)
b = torch.zeros(1, requires_grad=True) 

# 定义一个损失函数
def squared_loss(y_hat, y):
    """均方损失"""
    return (y_hat - y.reshape(y_hat.shape)) ** 2 / 2

# 定义优化算法
def sgd(params, lr, batch_size):
    """小批量随机梯度下降"""
    with torch.no_grad():
        for param in params:
            param -= lr * param.grad / batch_size
            param.grad.zero_()

# 开始学习
lr = 0.03           # 学习率
num_epochs = 3      # 迭代轮数
net = linreg        # 模型
loss = squared_loss # 损失函数

for epoch in range(num_epochs):     # 迭代数据集
    for X, y in data_iter(bathch_size, features, labels):       # 迭代批量
        l = loss(net(X, w, b), y) # 计算损失
        # 因为l形状是(batch_size, 1)，而不是一个标量
        # 所以sum()函数会返回一个标量
        l.sum().backward()
        sgd([w, b], lr, bathch_size) # 优化模型参数
    with torch.no_grad():
        train_l = loss(net(features, w, b), labels)
        print(f'epoch {epoch + 1}, loss {float(train_l.mean()):f}')
```
